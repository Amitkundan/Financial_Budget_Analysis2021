```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\user\Desktop\project\Financial_Budget_Analysis\India_budget_2021.csv")
data.head()
```

```python
data.shape
```

```python
# finding NaN values in the dataset
data.isnull().sum()
```

```python
data=data.dropna()
```

```python
data.isnull().sum()
```

```python
# Selecting rows within the valid range of the DataFrame
data = data.iloc[[0, 1, 2, 3, 4, 5, 6, 7, 8], :]

# Creating a new DataFrame for the row to append
new_row = pd.DataFrame({'Department /Ministry': ['OTHERS'], 'Fund allotted(in ₹crores)': [592971.0800000001]})

# Appending the new row to the DataFrame
data = pd.concat([data, new_row], ignore_index=True)

# Printing the modified DataFrame
print(data)
```

```python
data.plot.bar(x='Department /Ministry', y='Fund allotted(in ₹crores)',color='green')
```

```python
# creating a donut plot for clear distribution of budgets in various department
df = data["Fund allotted(in ₹crores)"]
labels = data["Department /Ministry"]
```

```python
plt.figure(figsize=(7,7))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85, shadow =True)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of The Budget", fontsize=20)
plt.show()
```
